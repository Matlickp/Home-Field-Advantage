```python
import pandas as pd
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.sportsDB
```

```python
df = pd.read_csv("MLB_wins_losses.csv")
df = df.rename(columns={"Team": "TEAM", "Year": "YEAR"})
df['HOME %'] = df['HOME %'] * 100
df['AWAY %'] = df['AWAY %'] * 100
df
```

```python
df_dict = df.to_dict(orient='records')
df_dict
```

```python
coll = db.mlbteamdata
```

```python
coll.remove()
```

```python
for record in df_dict:
    coll.insert_one(record)
```
