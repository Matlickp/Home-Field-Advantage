```python
import pandas as pd
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.sportsDB
```

```python
df = pd.read_json("mlb_stadium.json")
df = df.drop(columns='address')
df = df.rename(columns={'lat': 'LAT', 'lng': 'LNG', 'team': 'TEAM'})
df
```

```python
df_dict = df.to_dict(orient='records')
df_dict
```

```python
coll = db.mlbcoords
coll2 = db.mlbiconcoords
```

```python
coll.remove()
coll2.remove()
```

```python
for record in df_dict:
    coll.insert_one(record)
    coll2.insert_one(record)
```
