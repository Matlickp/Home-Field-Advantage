```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
```

```python
nba_teams = teams.get_teams()
nba_teams
```

```python
team_abbv = [team["full_name"] for team in nba_teams]
team_abbv = sorted(team_abbv)
team_abbv
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder()
```

```python
games = gamefinder.get_data_frames()[0]
games = games[games['TEAM_NAME'].isin(team_abbv)]
games.head(50)
```

```python
games = games.sort_values(by=['TEAM_NAME', 'GAME_DATE'])
games
```
