```python
import pandas as pd
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.sportsDB
```

```python
df = pd.read_csv('csv/games_percentage.csv')
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
p_df = pd.read_csv('csv/games_playoff_percentage.csv')
p_df = p_df.drop(columns='Unnamed: 0')
p_df.head()
```

```python
p_df_dict = p_df.to_dict(orient='records')

df_dict = df.to_dict(orient='records')
df_dict
```

```python
coll = db.nbateamdata
p_coll = db.nbaplayoffteamdata
c_coll = db.nbacoords
t_coll = db.nbamatchup
```

```python
for record in df_dict:
    coll.insert_one(record)

for record in p_df_dict:
    p_coll.insert_one(record)
```

```python
for record in coll.find():
    print(record)
```

```python
for record in p_coll.find():
    print(record)
```

```python
coords_df = pd.read_csv('csv/nba_state_coords.csv')
coords_df = coords_df.drop(columns='Unnamed: 0')
coords_df
```

```python
coords_dict = coords_df.to_dict(orient='records')

for record in coords_dict:
    c_coll.insert_one(record)
```

```python
winloss_df = pd.read_csv('csv/supercsv.csv')
winloss_df = winloss_df.drop(columns='Unnamed: 0')
winloss_df
```

```python
winloss_dict = winloss_df.to_dict(orient='records')

for record in winloss_dict:
    t_coll.insert_one(record)
```
