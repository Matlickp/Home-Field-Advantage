```python
import pandas as pd
import numpy as np
from nba_api.stats.static import teams
from datetime import datetime as dt
```

```python
nba_teams = teams.get_teams()
team_names = [team["full_name"] for team in nba_teams]
team_names = sorted(team_names)
team_names
```

```python
games_df = pd.read_csv("csv/games_ordered.csv")

games_df['SEASON_ID'] = games_df['SEASON_ID'].astype('str')
games_df['SEASON_ID'] = games_df['SEASON_ID'].str[1:]
games_df['SEASON_ID'] = games_df['SEASON_ID'].astype('int64')
```

```python
games_reg_season_df = games_df.loc[(games_df['GAME_ID'] < 40000000) & (games_df['GAME_ID'] >= 20000000)]
games_playoff_season_df = games_df.loc[(games_df['GAME_ID'] >= 40000000) & (games_df['GAME_ID'] < 50000000)]
games_reg_season_df['SEASON_ID'].value_counts()
```

```python
grouped_df = games_reg_season_df.groupby(['SEASON_ID', 'TEAM_NAME','LOCATION', 'WL'])
grouped_df.count()
```

```python
games_reg_season_df.to_csv('csv/games regular season.csv')
games_playoff_season_df.to_csv('csv/games playoffs.csv')
```

```python
games_df.dtypes
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017]

name_list = []
year_list = []
away_win_list = []
away_loss_list = []
home_win_list = []
home_loss_list = []
home_win_percentage_list = []
away_win_percentage_list = []


for team in team_names:
    print(team)
    for year in years:
        away_wins = 0
        away_loss = 0
        home_wins = 0
        home_loss = 0
        
        for index, row in games_reg_season_df.iterrows():

#             print(row["LOCATION"])
            if row["TEAM_NAME"] == team and row["SEASON_ID"] == year:

                if row["LOCATION"] == "HOME":

                    if row["WL"] == "W":
                        
                        home_wins += 1
                        
                    elif row["WL"] == "L":
                        
                        home_loss += 1
                        
                if row["LOCATION"] == "AWAY":
                    
                    if row["WL"] == "W":
                        
                        away_wins += 1
                    
                    elif row["WL"] == "L":
                        
                        away_loss += 1
        try:        
            home_win_percentage = home_wins / (home_wins + home_loss)
            away_win_percentage = away_wins / (away_wins + away_loss)

            name_list.append(team)
            year_list.append(year)
            away_win_list.append(away_wins)
            away_loss_list.append(away_loss)
            home_win_list.append(home_wins)
            home_loss_list.append(home_loss)
            home_win_percentage_list.append(home_win_percentage)
            away_win_percentage_list.append(away_win_percentage)
        except:
            print(f"no data for {team} year {year}")

        
```

```python
percentage_df = pd.DataFrame({
    'TEAM': name_list,
    'YEAR': year_list,
    'HOME W': home_win_list,
    'HOME L': home_loss_list,
    'AWAY W': away_win_list,
    'AWAY L': away_loss_list,
    'HOME %': home_win_percentage_list,
    'AWAY %': away_win_percentage_list
    
})
percentage_df.head()
```

```python
percentage_df.to_csv("csv/games_percentage.csv")
```
