```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.sportsDB
```

```python
df_dict = df.to_dict(orient='records')
```

```python
def fib(n):
    print('called fib')
    if(n <= 2):
        return 1
    return fib(n-1) + fib(n-2)

fib(10)
```
