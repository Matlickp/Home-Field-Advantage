```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
games_file = pd.read_csv("../Data/NFL/nfl_games.csv")
games_file.rename(index=str, columns={'home' : 'team1', 'away' : 'team2'}, inplace=True)
games_file.tail()
```

```python
home_group = games_file.rename(index=str, columns={'team1':'team'})
home_group = home_group.groupby(['season', 'team', 'result'], as_index=False).count()
home_group.drop(columns=['score1', 'score2', 'team2', 'lat', 'long'], inplace=True)
home_group.head()
```

```python
home_group.to_csv("../Data/NFL/grouped_results_home.csv")
```

```python
away_group = games_file.rename(index=str, columns={'team2':'team'})
away_group = away_group.groupby(['season', 'team', 'result'], as_index=False).count()
away_group.drop(columns=['score1', 'score2', 'team1', 'lat', 'long'], inplace=True)
away_group.head()
```

```python
away_group.to_csv("../Data/NFL/grouped_results_away.csv")
```
